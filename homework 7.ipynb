```python
import numpy as np
import pandas as pd
```

```python
#1.1--First use pd.read_cvs() function to load the dataset from the given file named global_air_quality.csv into a Pandas DataFrame.
#Print the DataFrame 
data = pd.read_csv("global_air_quality.csv")
data
```

```python
#1.2--The FactNumeric column contains the air quality information we want.
#With the values from the FactValueNumeric column, add a new column to your DataFrame called PM25_Value.
#Print the DataFrame containing only the FactValueNumeric and PM25_Value columns. 
#The should have identical numbers. 
data['PM25_Value'] = data['FactValueNumeric']
data
```

```python
print(data[['FactValueNumeric', 'PM25_Value']])
```

```python
# Group the countries by their ParentLocation
Location = data.groupby('ParentLocation')

# Calculate the average PM2.5 concentration for each continent and create a new DataFrame with these averages.
average_pm25 = Location['FactValueNumeric'].mean().reset_index()

# Rename the column from PM25_Value to Average PM25
average_pm25.columns = ['Continent', 'Average PM2.5']

# Add the PM25 Value column (assuming you're using 'FactValueNumericLow' for PM25 values)
data['PM25 Value'] = data['FactValueNumeric']

# Merge the average values back into the original DataFrame
new_avg_data = data.merge(average_pm25, left_on='ParentLocation', right_on='Continent', how='left')

# Print the DataFrame with to show columns Locations, Parent Location,PM25 Value and Average PM25.
result = new_avg_data[['ParentLocation', 'PM25 Value', 'Average PM2.5']]
print(result)
```

```python
# 1.4 Determine which continent (ParentLocation) has the highest average air pollution (concentrations of fine particular matter). 
#Does this surprise you? Why or why not? Include these answer as a comment(s) in your code.
max_avg_pm25 = average_pm25.loc[average_pm25['Average PM2.5'].idxmax()]
print(max_avg_pm25)
print(f"The continent with the highest average PM2.5 concentration is {max_avg_pm25['Continent']}")
print(f"It is not a surprise because the {max_avg_pm25['Continent']} has the highest average PM2.5")
```

```python
#Lastly, save the updated DataFrame with just the columns Locations, ParentLocation,PM25 Value and Average PM25 as a new CSV file to your system.
#Please submit this CSV file alongside your Jupyter Notebook submission for this assignment.
# Select the required columns
updated_data = new_avg_data[['Location', 'ParentLocation', 'PM25 Value', 'Average PM2.5']]

# Save the updated DataFrame as a new CSV file
updated_data.to_csv('updated_air_quality.csv', index=False)
```

```python
#2.1--First, load the dataset using sns.load dataset(‘planets’) and store it in a Pandas DataFrame. 
#Your task is to create a scatter plot that visualizes the relationship between orbital period (x-axis) and mass (y-axis) of all the exoplanets in the planets dataset. 
#Color the data points according to the exoplanet’s discovery method. 
#Try using seaborn instead of matplotlib to generate this plot
import seaborn as sns
import matplotlib.pyplot as plt

print(planets)
planets = sns.load_dataset('planets')

plt.figure(figsize=(12, 8))
sns.set(style="whitegrid", palette="pastel")

x = sns.scatterplot(data=planets, x='orbital_period', y='mass', hue='method', palette='deep')

x.set(xscale="log", yscale="log")

x.set_title('Orbital Period vs Mass', fontsize=13)
x.set_xlabel('Orbital Period (days)', fontsize=11)
x.set_ylabel('Mass(Earth mass)', fontsize=11)

x.legend_.set_title('Method')  
x.legend_.set_frame_on(True)  
x.legend_.prop.set_size(10)  

plt.show()
```

```python
#2.2 -- Next, create a bar chart where each stacked bar represents the total number of exoplanets discovered (y-axis) in each year (x-axis), grouped by the discovery method. 
#For example, a bar for 2024 would have sections for radial velocity,mircolensing, imaging, etc. 
#Each discovery method should have its own color.
#We highly recommend using seaborn over matplotlib for this bar chart.
#Before you submit this plot, make sure it has:
#• A descriptive title
#• Axis labels (with units!)
#• A legend explaining the exoplanet color coding

planets = planets.dropna(subset=['year', 'method'])

planets_discovered = planets.groupby(['year', 'method']).size().reset_index(name='count')

plt.figure(figsize=(16, 8))

sns.barplot(data=planets_discovered, x='year', y='count', hue='method', errorbar=None, palette='Set1')

plt.title('Exoplanet Discoveries vs years', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Number of exoplanets', fontsize=15)


plt.legend(title='Method', loc='upper left')

plt.show()
```
