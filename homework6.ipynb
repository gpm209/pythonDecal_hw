```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Part 1 
#1--write a function that takes three arguments (x data, an amplitude constant A, and a constant w for determining wavelength) 
#and returns a sinusoid

def sinosoid (x, A, w):
    return A * np.sin(w * x)

#2 --create an array to be your x data from 0 to 2π of 1000 points
x_data = np.linspace(0, 2 * np.pi, 1000)

#3-- create an array of 5 different amplitude A values, and a separate array of 5 different w values
A_values = np.array([1,2,3,4,5])
w_values = np.array([1,2,3,4,5])

#5 create a figure object to plot your graphs onto

fig, axs = plt.subplots()


#4 -- loop through each pair of A and w values. 
#For each iteration, calculate the array of y data that correspond to those values, using the same x data array for each iteration.
#Plot each array of y data onto your figure (Hint: you can plot at the same time you iterate)
for i in range(len(A_values)):
    
    y_data = A_values[i] * np.sin(w_values[i] * x_data)
    
    plt.plot(x_data, y_data, label=f'sine function {i}')

plt.xlabel("x data")
plt.ylabel("y data")
plt.title("5 Different Sine Functions")
plt.legend()
plt.legend(loc="lower left")
plt.show()
```
